```python
'''
Created on Jan 30, 2019

@author: canivel
'''

import os
import boto3
import sys

s3 = boto3.resource('s3')
BUCKETNAME = 'meracdrs.usmatrix.com'

def sendToS3(localpath, fileName, bucketName):
    return s3.Object(bucketName,fileName).upload_file(Filename=localpath)

def sendFiles(inputDirectory):
    try:
        directoryListing = os.listdir(inputDirectory)
        for filename in directoryListing:
            if filename.endswith(".gz"):
                localpath = inputDirectory + os.sep + filename
                print ("start Transfer {} to S3 bucket {}".format(filename, BUCKETNAME))
                sendToS3(localpath, filename, BUCKETNAME)
                print ("Finish Transfer to S3")
                print ("Removing local file {}".format(filename))
                os.remove(inputDirectory + os.sep + filename)
                print ("File {} removed".format(inputDirectory + os.sep + filename))
            
    except Exception as e:
        print("Directory error: " + str((e)))    
        
def main():
    inputDirectory = "/data/meralogs"
    sendFiles (inputDirectory)

if __name__ == '__main__':
    pid = str(os.getpid())
    pidfile = "/data/sendtos3.pid"

    if os.path.isfile(pidfile):
        print ("{} already exists, exiting".format(pidfile))
        sys.exit()
        
    open(pidfile, 'w').write(pid)
    try:
        main()
    finally:
        os.unlink(pidfile)
```
